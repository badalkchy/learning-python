# Function

```python
def fun():
  print("Hello from a function")
```

```python
fun()
```

```python
def my_function(fname): # definition
  print(fname + " Aera")

my_function("Amit") # call
```

```python
def my_function(fname, age): # definition
  print(fname + " Aera")
  print('I am ' + str(age) + ' year old')

my_function("Amit", 21) # call
```

```python
my_function("Vicky", 22) # call
```

```python
def myself(fname, lname, age): # definition
  print(fname + ' ' + lname)
  print('I am ' + str(age) + ' year old')

myself("Vicky", "Kumar", 22) # call
```

```python
str(22)
```

```python
def default(fname = 'Amit', age = 21): # definition
  print(fname)
  print('I am ' + str(age) + ' year old')

default("Vicky", 22) # call
```

```python
default()
```

```python
def my_function(*kids):
  print(type(kids))
  print("The youngest child is " + kids[2])

my_function("Emil", "Tobias", "Linus")

# * parameter used
```
